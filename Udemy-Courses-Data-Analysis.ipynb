```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(x=data['subject'])
plt.xlabel("Subjects", fontsize=13)
plt.ylabel("No. of Courses",fontsize=13)
plt.xticks(rotation=40)
plt.show()
```

```python
data["level"].value_counts()
```

```python
sns.countplot(x=data['level'])
plt.xlabel("Level", fontsize=13)
plt.ylabel("No. of Courses",fontsize=13)
plt.xticks(rotation=30)
plt.show()
```

```python
# Count of free and paid courses

data["is_paid"].value_counts()
```

```python
sns.countplot(x=data['is_paid'])
plt.xlabel("Paid", fontsize=13)
plt.ylabel("No. of Courses",fontsize=13)
plt.show()
```

```python
# Which course has more lectures(free or paid)

data.groupby(["is_paid"]).mean() 
```

```python
sns.barplot(x="is_paid", y="num_lectures", data=data)
plt.show()
```

```python
# Course with higher no. of subscribers(free or paid)

sns.barplot(x="is_paid", y="num_subscribers", data=data)
plt.show()
```

```python
# Level with highest no. of subscribers

sns.barplot(x="level", y="num_subscribers", data=data)
plt.show()
```

```python
# Most popular course title

data[data["num_subscribers"].max()==data["num_subscribers"]] ["course_title"]

```

```python
# 10 most popular courses as per no. of subscribers

top10=data.sort_values(by="num_subscribers", ascending=False).head(10)
```

```python
sns.barplot(x="num_subscribers", y="course_title", data = top10)
plt.show()
```

```python
# Subject with the highest no. of reviews

plt.figure(figsize=(8,5))
sns.barplot(x="subject", y="num_reviews" , data= data)
plt.show()
```

```python
# Does price affect no. of reviews

sns.scatterplot(x="price", y="num_reviews", data=data)
plt.show()
```

```python
# Total no. of courses related to python

len(data[data["course_title"].str.contains("python", case = False)])
```

```python
# 10 most popular python courses as per no. of subscribers

python10=data[data["course_title"].str.contains("python", case = False)].sort_values(by="num_subscribers", ascending=False).head(10)
```

```python
sns.barplot(x="num_subscribers", y="course_title", data= python10)
plt.show()
```

```python
# Year in which the highest no. of courses were posted

data["Year"]=data["published_timestamp"].dt.year
```

```python
sns.countplot(x="Year",data=data)
plt.show()
```

```python
# Category-wise count of posted subjects(year wise)

data.groupby("Year")["subject"].value_counts()
```
